```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('fight-with-covid-19-challenge/world_data.csv')
df1
```

```python
df1.isnull().sum()
```

```python
df1.drop(['NewCases', 'NewDeaths', 'NewRecovered'], axis = 1, inplace=True)
```

```python
df1.isnull().sum()
```

```python
df = pd.read_csv('fight-with-covid-19-challenge/covid_19_complete.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Province/State'], axis = 1, inplace=True)
```

```python
df
```

```python
df['Current'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
```

```python
df
```

```python
df_new = df.groupby('Date')[['Confirmed', 'Deaths', 'Recovered', 'Current']].sum()
```

```python
df_new
```

```python
plt.figure(figsize = [20, 15])
plt.plot(df_new['Confirmed'], color = 'g', label = 'Confirmed')
plt.plot(df_new['Deaths'], color = 'r', label = 'Deaths')
plt.plot(df_new['Recovered'], color = 'b', label = 'Recovered')
plt.plot(df_new['Current'], color = 'k', label = 'Current')
plt.legend()
```
